```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('SBM20_pulse_ADC900.txt', delimiter=',')

plt.figure()
x = data[:,0]
y = data[:,1]
logy = np.log(y)
#plt.plot( x, logy )

p0,p1 = np.polyfit(x,logy,1)
print(1/p0,p1)

cx = np.linspace(0,1000,1000)

plt.plot(x,logy)
plt.plot(cx, p0*cx+p1 )

plt.figure()
plt.plot(x, y)
plt.plot(cx, np.exp(p0*cx+p1) )

print("Peak voltage is", np.exp(p1), "V")
```
